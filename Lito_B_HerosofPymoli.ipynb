```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
csvpath = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(csvpath)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
Players = len(purchase_data["SN"].unique())

totalplayers_df = pd.DataFrame([{"Total Players": Players}])
totalplayers_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = purchase_data["Item ID"].value_counts()
items

uniques = pd.DataFrame(purchase_data["Item ID"].unique())
len(uniques)


drop_dupes_i = purchase_data.drop_duplicates(["Item ID"], keep = "last")
drop_dupes_i

total_uniques = len(drop_dupes_i)

total_purchases = purchase_data["Price"].count()
total_revenue = round(purchase_data["Price"].sum(),2)

average_p = round(total_revenue/total_purchases, 2)

purchases = pd.DataFrame([{ "Number of Unique Items": total_uniques,
                            "Average Price": average_p,
                            "Number of Purchases": total_purchases,
                            "Total Revenue": total_revenue}])

purchases = purchases[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

purchases
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
drop_dupes_g = purchase_data.drop_duplicates(["SN"], keep="last")

g_demograph = drop_dupes_g["Gender"].value_counts().reset_index()
g_demograph

g_demograph["Percentage of Players"] = g_demograph["Gender"]/Players * 100

g_demograph.rename(columns = {"index": "Gender", "Gender": "Number of Players"}, inplace= True)
g_demograph.set_index(["Gender"], inplace = True)

g_demograph
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = pd.DataFrame(purchase_data.groupby("Gender")["Gender"].count())

total_gen_purchase = pd.DataFrame(purchase_data.groupby("Gender")["Price"].sum())

purchase_analysis = pd.merge(purchase_count, total_gen_purchase, right_index = True, left_index = True)

purchase_analysis.rename(columns = {"Gender": "Purchase Count", "Price": "Total Purchase Value"}, inplace = True)

purchase_analysis["Average Purchase Price"] = purchase_analysis["Total Purchase Value"]/purchase_analysis["Purchase Count"]

purchase_analysis = purchase_analysis.merge(g_demograph, left_index = True, right_index = True)

purchase_analysis["Avg Total Purchase per Person"] = purchase_analysis["Total Purchase Value"]/purchase_analysis["Number of Players"]

del purchase_analysis["Percentage of Players"]
del purchase_analysis["Number of Players"]

purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bin_age = [0, 9.90, 14.90, 19.90, 24.9, 29.9, 34.90, 39.90, 99999999999]
age_group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

age_demo = purchase_data
age_demo["Age Summary"] = pd.cut(age_demo["Age"], bin_age, labels= age_group)

ages = purchase_data.groupby("Age Summary").SN.nunique().reset_index(name="Total Count")

percentage = round((ages["Total Count"] /Players)*100,2)
ages["Percentage of Players"] = percentage
ages
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchases_byage = purchase_data.groupby(["Age Summary"])
price_byage = purchases_byage["Price"].sum()
purchase_byage = pd.DataFrame(price_byage)
purchase_byage = purchase_byage.rename(columns={"Price":"Total Purchase Value"})

count_byage = purchases_byage["Age Summary"].count()
purchase_byage["Purchase Count"] = pd.DataFrame(count_byage)

count_byage     = purchases_byage["SN"].nunique()
purchase_byage["Total Count"]  = pd.DataFrame(count_byage)

purchase_byage["Average Purchase Price"]=round(purchase_byage["Total Purchase Value"]/purchase_byage["Purchase Count"] ,2).sort_values(ascending=True)
purchase_byage["Avg Total Purchase per Person"]=round(purchase_byage["Total Purchase Value"]/purchase_byage["Total Count"],2).sort_values(ascending=True)

purchase_byage=purchase_byage[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]

purchase_byage
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchases_byspender = purchase_data.groupby(["SN"])

price_byspender     = purchases_byspender["Price"].sum().sort_values(ascending=False)

purchase_byspender  = pd.DataFrame(price_byspender)

purchase_byspender  = purchase_byspender.rename(columns={"Price":"Total Purchase Value"})

purchase_byspender.sort_values(["Total Purchase Value"], ascending=False)

spender_count=purchase_data.groupby("SN").count()

spender_Detail  = pd.DataFrame(spender_count)

purchase_byspender["Purchase Count"] = spender_Detail["Purchase ID"]

purchase_byspender["Average Purchase Price"]= round((purchase_byspender["Total Purchase Value"]/purchase_byspender["Purchase Count"]),2)

purchase_byspender=purchase_byspender[["Purchase Count","Average Purchase Price","Total Purchase Value"]]
purchase_byspender.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchases_byitem = purchase_data.groupby(["Item ID","Item Name"])

price_byitem     = purchases_byitem["Price"].sum().sort_values(ascending=False)

purchases_byitem  = pd.DataFrame(price_byitem)

purchases_byitem  = purchases_byitem.rename(columns={"Price":"Total Purchase Value"})

Item_count=purchase_data.groupby(["Item ID","Item Name"])

Item_Detail=Item_count.count()

purchases_byitem["Purchase Count"] = Item_Detail["Purchase ID"].astype('int')

avg_item_price  =  purchase_data.groupby(["Item ID","Item Name"]).mean()["Price"]

purchases_byitem["Item Price"] = pd.DataFrame(avg_item_price)

purchases_byitem = purchases_byitem[["Purchase Count","Item Price","Total Purchase Value"]]

purchases_byitem.sort_values(["Purchase Count"], ascending=False)

purchases_byitem
```
